## EDA(EXPLORATORY DATA ANALYSIS)


```python
ax = sns.countplot(x='Gender',data =df, hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
dx = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
import matplotlib.pyplot as plt
sns.barplot(x='Age Group',y='Amount',data=dx,hue='Age Group',palette='dark')
```

```python

ax = sns.countplot(x='Age Group',data=df, hue='Gender',palette='dark')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python

ss = df.groupby(['State'])['Orders'].sum().sort_values(ascending=False).head(10)
```

```python
sns.barplot(x='State',y='Orders',data=ss.reset_index(),hue='State')
plt.figure(figsize=(15,10))
```

```python

plt.figure(figsize=(10,6))
sns.countplot(x='Marital_Status',data=df,palette='dark')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
plt.figure(figsize=(19,6))
sns.countplot(x='Occupation',data=df,palette='dark')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
plt.figure(figsize=(25, 6))
top_10_categories = df['Product_Category'].value_counts().iloc[:10].index
ax = sns.countplot(
    x='Product_Category',
    data=df[df['Product_Category'].isin(top_10_categories)],
    palette='dark',
    order=top_10_categories
)
```
